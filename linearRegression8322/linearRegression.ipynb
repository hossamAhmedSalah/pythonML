```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split as TSP 
#let us have the csv file 
salary = pd.read_csv('Salary_Data.csv')
years = salary['YearsExperience']
salary2 = salary['Salary']
plt.plot(years,salary2, 'ro') #''r.' 'ro-' 'ro--'
plt.title("relation between years of experience and Salary")
plt.xlabel("years of experience")
plt.ylabel("Salary")
plt.figure(figsize=(7,5), dpi=300)
plt.show()
#plt.scatter(years, salary2, color = 'purple')
#plt.show()

```

```python
 #test train split
 years_train, years_test, salary2_train, salary2_test = TSP(years, salary2, test_size=1/3)
 #test_train_split visualy
 plt.plot(years, salary2, 'rx',markersize=10, label='all data set')
 plt.plot(years_train, salary2_train, 'yo', label='Training Data')
 plt.plot(years_test, salary2_test, 'bo', label='tested data')
 plt.legend()
 plt.show()
```

```python
#create the linear model 
LR.fit(years_train.values.reshape(-1,1), salary2_train.values)
```
