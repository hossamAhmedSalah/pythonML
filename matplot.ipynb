```python
import matplotlib.pyplot as pit
import numpy as np
import pandas as pd
```

```python
#pit.plot([1,2,0,3], [2,3,4,7])
x = [1,2,3,4]
y = [5,6,0,8]

pit.plot(x,y , label='someRandomGraph', color='red', linewidth=9, marker='X',markersize=15, markeredgecolor='yellow',linestyle='--')
pit.title("HOSSAM Ahmed")
pit.xlabel("HI you x")
pit.ylabel("yyyyou")
pit.xticks([0, 1, 2, 3, 4,5])
pit.yticks([1,3,5,7,10,20,40])
pit.legend()
pit.show()
```

```python
#shorthand way 
pit.figure(figsize=(5,3), dpi=150)
pit.plot(x,y, 'rx--', label='x') # [color][marker][line]
x2 = np.arange(0,4,0.5)
print(x2)
pit.plot(x2, x2**2.55, label='x^2.55', color='green',linestyle='--', marker='o')
pit.legend()
pit.savefig('graphy82.png', dpi=3000)
pit.show()
```

Real world examples

```python
gas = pd.read_csv('gas_prices.csv')
print(gas.head())
pit.figure(figsize=(7,3), dpi= 200)
pit.title("Gas prices in some countries")
pit.xlabel('Year')
pit.ylabel('price')
pit.plot(gas.Year, gas.USA ,  '.-', label='USA')
pit.plot(gas.Year, gas.Canada, 'x-', label='Canada')
pit.plot(gas.Year, gas.France , 'o-', label='France')
pit.plot(gas.Year, gas.Australia, '--', label='Australia')
pit.plot(gas.Year, gas.Germany, label='Germany')
pit.plot(gas.Year, gas.Italy, label='Italy')
pit.legend()
pit.xticks(gas.Year[::2])


pit.show()
```

```python
for country in gas: 
    if country == 'Year':
        continue
    print(country)
    pit.plot(gas.Year, gas[country])
```

```python
#fifa data #this data sound to be  old it had been changed alot thought
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
#HISTOGRAMS
bins = [40,50,60,70,80,90,100]

pit.hist(fifa.Overall, bins=bins)
pit.xlabel("skill level")
pit.ylabel("number of player")
pit.title("player skills in fifa 2018")
pit.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
pit.pie([left, right], labels=['left','right'],autopct='%.2f %%')
pit.title("prefernce of FOOT")
pit.legend()
pit.show()
```
